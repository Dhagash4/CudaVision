```python
import os
import numpy as np
from skimage import io
import skimage
import skimage.transform
```

```python
root_dir = "data_augmentation/images"
annotation_path = "data_augmentation/trainval.txt"

data = np.loadtxt(annotation_path,dtype=str)
print(data.shape)
```

```python
labels = []
```

```python
for i in range(data.shape[0]):
    img_save = data[i,0].split(".")[0]
    img_pth = os.path.join(root_dir,data[i,0])
    img = io.imread(img_pth)
    select_saptial_augmentation = np.random.randint(0,2,dtype=int)
    select_color_augmentation = np.random.randint(0, 2, dtype=int)
    
    labels.append([data[i, 0], data[i, 1]])

    
    #Rotate
    rotate = np.random.randint(10,181,dtype=int)
    img_rotate = skimage.transform.rotate(image=img,angle=rotate,mode='wrap')
    io.imsave(os.path.join(root_dir,img_save + "_rotate" + ".jpg"),img_rotate)
    labels.append([img_save + "_rotate" + ".jpg",data[i,1]])
    #Flipping
    img_flip_horizontal = img[:, ::-1]
    io.imsave(os.path.join(root_dir, img_save + "_fliphorizontal" +
              ".jpg"), img_flip_horizontal)
    labels.append([img_save + "_fliphorizontal" + ".jpg", data[i, 1]])

   
    img_flip_vertical = img[::-1, :]
    io.imsave(os.path.join(root_dir, img_save + "_flipvertical" +
              ".jpg"), img_flip_vertical)
    labels.append([img_save + "_flipvertical" + ".jpg", data[i, 1]])
    
    #Scaling
    img_scale = skimage.transform.rescale(img, scale=0.3, mode='edge')
    io.imsave(os.path.join(root_dir, img_save + "_scale" + ".jpg"),
              img_scale)
    labels.append([img_save + "_scale" + ".jpg", data[i, 1]])
    
    #Gaussian Noise

    img_noise = skimage.util.random_noise(img,mode='gaussian')
    io.imsave(os.path.join(root_dir, img_save + "_noise" + ".jpg"),
              img_noise)
    labels.append([img_save + "_noise" + ".jpg", data[i, 1]])



       
```
